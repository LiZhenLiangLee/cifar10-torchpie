```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import torchvision.transforms as T
from PIL import Image
```

```python
torch_transform = T.Compose([
    T.RandomCrop(32, padding=4),
    T.RandomHorizontalFlip(),
])
```

```python
with open('data/cifar-10-batches-py/data_batch_1', 'rb') as f:
    cifar10 = pickle.load(f, encoding='latin1')
```

```python
np_image = cifar10['data'][0].reshape(3, 32, 32).transpose(1, 2, 0)
```

```python
plt.imshow(np_image)
```

```python
pil_image = Image.fromarray(np_image)
```

```python
plt.imshow(pil_image)
```

```python
pil_image_transformed = torch_transform(pil_image)
plt.imshow(pil_image_transformed)
```

```python
np_image_padded = np.pad(np_image, ((4, 4), (4,4), (0,0)), 'constant')
plt.imshow(np_image_padded)
```

```python
def crop(image, i, j, h, w):
    return image[i: i + h, j: j+w]
```

```python
np_image_cropped = crop(np_image_padded, 6, 1, 32, 32)
plt.imshow(np_image_cropped)
```

```python
np_image_from_pil = np.asarray(pil_image_transformed)
plt.imshow(np_image_from_pil)
```

```python
np.array_equal(np_image_cropped, np_image_from_pil)
```

```python
plt.imshow(np.fliplr(np_image_cropped))
```
